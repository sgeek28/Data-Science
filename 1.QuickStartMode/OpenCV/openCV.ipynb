```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('/home/sneha/Downloads/DataScience-CodingBlocks/QuickStartMode/OpenCV/bulbasaur.jpg')
rgb_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
print(img.shape)
#crop image rows,columns,channels
#we are keeping all channels :
crop_img=rgb_img[100:450,0:450,:]
plt.imshow(crop_img)
#turn off axis
plt.axis("off")
plt.show()

print(crop_img.shape)
```

```python
total_pixels=crop_img.shape[0]*crop_img.shape[1]
print(total_pixels)
#each pixel has 3 channels
#157500*3 integers to store one image
#it is always suggested to use less number of pixels

#resizing image to lower pixels
resize_img=cv2.resize(crop_img,(100,100))
plt.imshow(resize_img)
plt.axis("off")
plt.title("Bulbasaur"+str(resize_img.shape))
plt.show()
```
